```python
import json

def convert(text):
    genres_list = []
    
    for item in json.loads(text):
        genres_list.append(item['name'])
        
    return genres_list
```

```python
print(convert(movies.iloc[0]['genres']))
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert_cast(text):
    cast_list = []

    for item in json.loads(text)[:3]:
        cast_list.append(item['name'])

    return cast_list
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies[['title', 'cast']].head()
```

```python
crew = json.loads(movies.iloc[0]['crew'])

print(crew[:5])
```

```python
def fetch_director(text):
    for item in json.loads(text):
        if item['job'] == 'Director':
            return [item['name']]

    return []
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies[['title', 'crew']].head()
```

```python
def remove_space(lst):
    new_list = []

    for item in lst:
        item = item.replace(" ", "")
        new_list.append(item)

    return new_list
```

```python
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
```

```python
movies[['genres', 'cast', 'crew']].head()
```

```python
movies['overview'] = movies['overview'].fillna('')
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies['tags'] = (
    movies['overview']
    + movies['genres']
    + movies['keywords']
    + movies['cast']
    + movies['crew']
)
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
```

```python
movies = movies[['movie_id', 'title', 'tags']]
```

```python
movies.head()
```

```python
movies['tags'] = movies['tags'].apply(lambda x: x.lower())
```

```python
movies.iloc[0]['tags']
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):

    stemmed_words = []

    for word in text.split():
        stemmed_words.append(ps.stem(word))

    return " ".join(stemmed_words)
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

```python
movies.head()
```

```python
movies.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(
    max_features=5000,
    stop_words='english'
)

vectors = cv.fit_transform(movies['tags'])

vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)

similarity.shape
```

```python
movies['title_lower'] = movies['title'].str.casefold()

def recommend(movie, top_n=5):

    matched_movies = movies[
        movies['title_lower'] == movie.casefold()
    ]

    if matched_movies.empty:
        print("❌ Movie not found.")
        return

    movie_index = matched_movies.index[0]

    distances = sorted(
        enumerate(similarity[movie_index]),
        reverse=True,
        key=lambda x: x[1]
    )

    for index, similarity_score in distances[1:top_n+1]:
        movie_title = movies.iloc[index]['title']

        print(f"Movie Title: {movie_title}, Similarity Score: {similarity_score:.3f}")

```

```python
recommend("Avatar")
```

```python
recommend("Avvattrrr")
```

```python
#Flow:
# Raw Data
#       │
#       ▼
# Preprocessing
#       │
#       ▼
# CountVectorizer
#       │
#       ▼
# vectors
#       │
#       ▼
# Cosine Similarity
#       │
#       ▼
# similarity

# Our recommondor system uses-
# movie name
#       │
#       ▼
# movie_index
#       │
#       ▼
# similarity[movie_index]
#       │
#       ▼
# Top recommendations
```
